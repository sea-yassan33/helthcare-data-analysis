```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import japanize_matplotlib
import seaborn as sns
```

```python
df = pd.read_csv('年次国民医療費.csv')
df
```

```python
## 国民医療費(億円)の年次推移
plt.figure(figsize=(10, 5))
sns.lineplot(x='年次', y='国民医療費(億円)', data=df)
plt.title('国民医療費の年次推移')
plt.show()
```

```python
# 国民医療費とGDPとの年次推移
# グラフのスタイル設定
fig, ax1 = plt.subplots(figsize=(10, 5))

# 1つ目のY軸：国民医療費の年次推移
bar=ax1.bar(df['年次'], df['国民医療費(億円)'], color='lightblue', label='国民医療費(億円)',alpha=0.7)
ax1.set_ylabel('国民医療費(億円)', fontsize=12)
ax1.set_ylim(0,500000)
ax1.set_xlabel('年次', fontsize=12)
## 左Y軸にカンマ区切りを適用
ax1.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f"{int(x):,}"))
# 2つ目のY軸：国民医療費/GDPの年次推移
ax2 = ax1.twinx()
## リストで返す為、カンマを付ける必要がある
line1,=ax2.plot(df["年次"], df["国民医療費/GDP(%)"], marker="o", linestyle="--", color="blue", label="国民医療費/GDP(%)")
ax2.set_ylabel("比率 (%)", fontsize=12)
ax2.set_ylim(2, 14)

# 凡例
#ax1.legend(loc="upper left", fontsize=10)
#ax2.legend(loc="upper right", fontsize=10)
## すべての凡例を1つにまとめ、左上に縦に並べる
fig.legend([bar, line1], ["国民医療費", "対国内総生産 (GDP) 比率"], loc="upper left", fontsize=10, ncol=1)

# タイトル
plt.title("国民医療費と経済指標の推移", fontsize=14)

plt.show()
```
