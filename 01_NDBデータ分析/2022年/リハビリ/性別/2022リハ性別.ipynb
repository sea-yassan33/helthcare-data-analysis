```python
df = pd.read_csv('./test.csv')
df = df[df["性別"] == "M"]
# 性別ごとに疾患別の推移を集計
diseases = df.columns[2:]  # 疾患のカラムを取得

# 年齢区分のマッピング
age_mapping = {
    1: "0～4歳", 2: "5～9歳", 3: "10～14歳", 4: "15～19歳", 5: "20～24歳",
    6: "25～29歳", 7: "30～34歳", 8: "35～39歳", 9: "40～44歳", 10: "45～49歳",
    11: "50～54歳", 12: "55～59歳", 13: "60～64歳", 14: "65～69歳", 15: "70～74歳",
    16: "75～79歳", 17: "80～84歳", 18: "85～89歳", 19: "90歳以上"
}
df["年齢区分"] = df["年齢区分"].map(age_mapping)

# 疾患ごとにプロット
diseases = ["心大血管疾患", "脳神経系疾患", "廃用症候群", "運動器疾患", "呼吸器疾患", "がん", "認知"]

plt.figure(figsize=(12, 6))
for disease in diseases:
    plt.plot(df["年齢区分"], df[disease], marker="o", label=disease)

plt.xticks(rotation=45)
plt.xlabel("年齢区分")
plt.ylabel("患者数")
plt.title("疾患別の年齢推移(男性)")
plt.legend()
plt.grid()
plt.show()

df = pd.read_csv('./test.csv')
df = df[df["性別"] == "F"]
# 性別ごとに疾患別の推移を集計
diseases = df.columns[2:]  # 疾患のカラムを取得
df["年齢区分"] = df["年齢区分"].map(age_mapping)

plt.figure(figsize=(12, 6))
for disease in diseases:
    plt.plot(df["年齢区分"], df[disease], marker="o", label=disease)

plt.xticks(rotation=45)
plt.xlabel("年齢区分")
plt.ylabel("患者数")
plt.title("疾患別の年齢推移(女性)")
plt.legend()
plt.grid()
plt.show()
```
