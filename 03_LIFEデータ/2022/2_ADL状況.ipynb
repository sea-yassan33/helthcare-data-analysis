```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import japanize_matplotlib
import seaborn as sns
```

```python
adl_list=['食事','移乗','整容','トイレ動作','入浴','歩行','階段','更衣','排尿','排便']
## 関数
def plot_service(df, service):
  df2 = df.copy().filter(like=f'{service}_')
  df2_ratio = df2.div(df2.sum(axis=1), axis=0) * 100
  df2_ratio.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='Set3')
  plt.title(f'{service}に対するADL状況の比率') 
  plt.xlabel('サービス種類')
  plt.ylabel('比率')
  plt.xticks(rotation=90, fontsize=10)
  plt.legend(title='ADL状況', bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.tight_layout()
  plt.show()
```

```python
df = pd.read_csv('./data/サービス種類別_ADL状況別.tsv', sep='\t',index_col=0)
df.head(2)
```

```python
df = pd.read_csv('./data/サービス種類別_ADL状況別.tsv', sep='\t',index_col=0)
for adl in (adl_list):
  plot_service(df,adl)
```
