```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import japanize_matplotlib
import seaborn as sns
```

```python
## 関数
def plot_service(df):
  df2 = df.copy()
  df2.drop(columns='総数', inplace=True)
  df2.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='Set3')
  plt.title(f'施設別LIFE登録状況') 
  plt.xticks(rotation=90, fontsize=10)
  plt.legend(title='LIFEt登録種別', bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.grid(axis='y', linestyle='--', alpha=0.3)
  plt.tight_layout()
  plt.show()
def plot_service2(df):
  df2 = df.copy()
  df2.drop(columns='総数', inplace=True)
  df2_ratio = df2.div(df2.sum(axis=1), axis=0) * 100
  df2_ratio.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='Set3')
  plt.title(f'施設別LIFE登録状況の比率') 
  plt.ylabel('比率')
  plt.xticks(rotation=90, fontsize=10)
  plt.legend(title='LIFEt登録種別割合', bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.grid(axis='y', linestyle='--', alpha=0.3)
  plt.tight_layout()
  plt.show()
```

```python
df = pd.read_csv('./data/サービス種類別_実事業所数.tsv', sep='\t',index_col=0)
df.head(2)
```

```python
plot_service(df)
```

```python
plot_service2(df)
```
